## Описание проекта

Цель исследования является разработка модели по предсказанию успешности стартапа.

## Предобработка данных

Как мы видели из информации о файлах, в них встречается довольно много пропусков.   
Оценим визуально картину, нет ли взаимосвязи между пропущенными значениями в данных?

```python
msno.matrix(data_train_kg);
```

```python
msno.heatmap(data_train_kg);
```

Получается, что между собой связаны пропуски в полях "state_code", "regoin", "country_code" и "city". Если не заполнено одно из этих полей - то с больщой долей вевроятности будет пропущено значение и в трех других полях.   
Что касается колонки "closed_at" - то ее пропуски объяснимы. Если данных в колонке нет - значит, стартап действущий, и все в порядке.   
Поскольку пропуски составляют порядка 10% данных, полагаю, будет правильным заполнить пропущенные значения "заглушкой" - "unknown".

```python
# оценим количество пропусков

msno.bar(data_train_kg);
```

Что касается пропусков в колонке "funding total usd" - можно предположить, что стартап не набралвообще никакой суммы. В этом случае пропущенные значения можно заменить нулевыми.    
Создадим копию нашего датафрейма. Произведем обозначенные замены.

```python
data_train = data_train_kg.reset_index(drop=True)
```

```python
data_train[['country_code', 'state_code', 'region', 'city']] = \
             data_train[['country_code', 'state_code', 'region', 'city']].fillna(value='unknown')
data_train.info()
```

```python
data_train['funding_total_usd'] = data_train['funding_total_usd'].fillna(0)
data_train.info()
```

У нас остались пропуски: 1 значение в колонке названия. Около 2,5 тысяч в колонке категорий. И отсутствующие даты в колонке "closed at".   
- отсутствующее название проще удалить
- пропуски в категории заменить на unknown
- пропуски даты в колонке заполнить на дату окончания выборки 2018-01-01

```python
data_train = data_train.dropna(subset=['name'])
data_train.info()
```

```python
data_train['category_list'] = data_train['category_list'].fillna(value='unknown')
data_train.info()
```

```python
data_train['closed_at'] = data_train['closed_at'].fillna(value='2018-01-01')
data_train['closed_at'] = pd.to_datetime(data_train['closed_at'], format= '%Y-%m-%d')
data_train.info()
```

```python
data_train['lifetime'] = (( data_train['closed_at'] - data_train['founded_at'] ) / np.timedelta64 ( 1 , 'D')).astype('int')
display(data_train.head())
data_train.info()
```

Поиск дубликатов: явные, неявные.

```python
# проверим на полные явные дубликаты

data_train.duplicated(keep=False).sum()
```

Явных дубликатов нет. Посмотрим на уникальные значения в каждой из колонок, кроме дат.

```python
def uniq (df, columns):
    for column in columns:
        print('Название колонки:', column)
        print('Количество уникальных значений:', df[column].nunique())
        print(df[column].value_counts())
        print()
```

```python
uniq(data_train, data_train.columns)
```

Весьма интересно выглядят данные по "дате основания" стартапа - значительное количество их зафиксировано для первых дней годов. Также мы видим, что значительная часть стартапов продолжает свое существование.   
Добавим колонку с годом основания стартапа для более наглядного графика распределения в следующем разделе исследования.

```python
# добавим колонку с годом создания стартапа

data_train['year_of_foundation'] = pd.DatetimeIndex(data_train['founded_at']).year
data_train['year_of_foundation'].value_counts()
```

Интересно, что в 2016 году зарегистрированы всего 2 новых стартапа.

#### Исследование географических параметров

Уже из раздела Предобработки данных было видно, что основная часть стартапов сконцентрирована в США. В настоящий момент исследователи географически разделяют данные по странам, городам, иногда объединяют данные по целым континентам. Так, например, значительным количеством стартапов могут похвататься Индия, Китай, а страны Африки исследователи иногда объединяют в одну группу.   
У нас есть данные по странам, регионам и городам.   
Посчитаем лидирующие позиции по этим параметрам, а также можно разделить страны по континентам, и посчитать топ-10 стран на каждом континенте.

В терминологии существует различие: материками считаются Евразия, Африка, Северная Америка, Южная Америка, Австралия и Антарктида. Однако, следует понимать, что историко-культурные различия между государствами Евразии велики, поэтому есть еще историко-культурное деление на "части света". К частям света относят: Азию, Африку, Америку, Европу, Австралию и Океанию, а также Антарктику. Пожалуй, в текущем исследовании мне ближе именно историко-кульруное деление.

```python
# создадим функцию для категоризации континента

def def_parts(df):
    if df in(['AUS', 'VUT', 'PNG', 'SLB', 'FJI', 'KIR', 'MHL', 'NRU',
                      'NZL', 'PLW', 'WSM', 'TON', 'TUV', 'FSM']):
        return 'Australia&Oceania'
    if df in(['AUT', 'BEL', 'GBR', 'DEU', 'IRL', 'LIE', 'LUX', 'MCO',
                      'NLD', 'FRA', 'CHE', 'BLR', 'BGR', 'HUN', 'MDA', 'POL',
                      'ROU', 'SVK', 'UKR', 'CZE', 'DNK', 'ISL', 'NOR', 'JEY',
                      'LVA', 'LTU', 'FIN', 'SWE', 'EST', 'ALB', 'AND', 'BIH',
                      'VAT', 'GRC', 'ESP', 'ITA', 'MLT', 'PRT', 'SMR', 'MKD',
                      'SRB', 'SVN', 'HRV', 'MNE', 'GIB', 'GGY']):
        return 'Europe'
    if df in(['DZA', 'EGY', 'ESH', 'LBY', 'MAR', 'SDN', 'TUN', 'BEN',
                      'BFA', 'GMB', 'GHA', 'GIN', 'GNB', 'CPV', 'CIV', 'LBR',
                      'MRT', 'MLI', 'NER', 'NGA', 'SEN', 'SLE', 'TGO', 'GAB',
                      'CMR', 'COD', 'COG', 'STP', 'CAF', 'TCD', 'GNQ', 'BDI',
                      'DJI', 'KEN', 'RWA', 'SOM', 'TZA', 'UGA', 'ERI', 'ETH', 
                      'SSD', 'AGO', 'BWA', 'ZMB', 'ZWE', 'COM', 'LSO', 'MUS', 
                      'MDG', 'MWI', 'MOZ', 'NAM', 'SWZ', 'SYC', 'ZAF']):
        return 'Africa'
    if df in(['AZE', 'ARM', 'AFG', 'BGD', 'BHR', 'BRN', 'BTN', 'TLS',
                      'VNM', 'GEO', 'ISR', 'IND', 'IDN', 'JOR', 'IRQ', 'IRN',
                      'YEM', 'KAZ', 'KHM', 'QAT', 'CYP', 'KGZ', 'CHN', 'HKG',
                      'MAC', 'TWN', 'PRK', 'KOR', 'KWT', 'LAO', 'LBN', 'MYS',
                      'MDV', 'MNG', 'MMR', 'NPL', 'ARE', 'OMN', 'PAK', 'PSE',
                      'RUS', 'SAU', 'SGP', 'SYR', 'TJK', 'THA', 'TKM', 'TUR',
                      'UZB', 'PHL', 'LKA', 'JPN']):
        return 'Asia'
    if df in(['CAN', 'MEX', 'USA', 'BLZ', 'GTM', 'HND', 'CRI', 'NIC',
                      'PAN', 'SLV', 'ATG', 'BHS', 'BRB', 'HTI', 'GRD', 'DMA',
                      'DOM', 'CUB', 'VCT', 'KNA', 'LCA', 'TTO', 'JAM', 'ARG',
                      'BOL', 'BRA', 'VEN', 'GUY', 'COL', 'PRY', 'PER', 'SUR', 
                      'URY', 'CHL', 'ECU', 'CYM', 'BMU', 'PRI', 'MAF', 'BLM']):
        return 'America'
    else:
        return 'other'
```

```python
# добавим колонку с частью света в тренировочную выборку

data_train['part_of_world'] = data_train['country_code'].apply(def_parts)
data_train['part_of_world'].value_counts()
```

## Поиск наилучшей модели

В наших данных все еще значительное количество уникальных значений параметров.    
- Для обучения модели выберем часть данных. 
- Разделим выборку на тренировочную и валидационную.
- Подготовим данные для обучения с помощью пайплайна.
- поищем через поиск наилучшую модель.
- выбранные признаки - 'funding_rounds','lifetime', 'month_of_foundation', 'time_to_fund', 'part_of_world'

```python
data_for_model = data_train_filt[['name', 'status','lifetime', 'part_of_world', 'month_of_foundation', 'time_to_fund']]
```

```python
num_cols = ['lifetime', 'month_of_foundation', 'time_to_fund']
ohe_cols = ['part_of_world']
```

```python
#

X_train, X_valid, y_train, y_valid = train_test_split(
    data_for_model.drop(
        ['name', 'status'],axis=1),
    data_for_model['status'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    stratify = data_for_model['status']
)
```

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import ADASYN
from imblearn.over_sampling import SMOTENC

sampler = SMOTENC(categorical_features = ['part_of_world'], random_state=42, k_neighbors=5)
X_train, y_train = sampler.fit_resample(X_train, y_train) 
X_train.shape[0]
```

```python
print(X_train.shape)
print(X_valid.shape)
print(y_train.shape)
y_valid.shape
```

### Подготовка данных

```python
label = LabelEncoder()
print(y_train)

y_train = label.fit_transform(y_train)
print(y_train)
y_valid = label.transform(y_valid)
```

```python
# создаем пайплайн для подготовки данных

ohe_pipe = Pipeline([('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))])


data_preprocessor = ColumnTransformer([
    ('ohe', ohe_pipe, ohe_cols),
    ('num', MinMaxScaler(), num_cols)
    ], 
    remainder = 'passthrough'
)

# создаём итоговый пайплайн

pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

### Поиск параметров моделей

```python
# задаем параметры перебора моделей и их гиперпараметров

param = [
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 10),
        'models__max_features': range(2, 10),
        'models__min_samples_leaf': range (2,10),
        'models__min_samples_split': range (2,10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler()]  
    },
    
    {
        'models': [RandomForestClassifier()],
        'models__n_estimators': range (3, 21, 3),
        'models__max_depth': range (4,10),
        'models__min_samples_leaf': range (1,10),
        'models__min_samples_split': range (2,10) ,
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler()]
    },

    #         {
    #     'models': [KNeighborsClassifier()],
    #     'models__n_neighbors': range(2, 6),
    #     'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler()], 
    # },

    # {
    #     'models': [LogisticRegression(
    #         random_state=RANDOM_STATE, 
    #         solver='liblinear'
    #     )],
    #     'models__C': range(1,5),
    #     'models__penalty' : ['l1', 'l2'],
    #     'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']  
    # }
]
```

```python
rand = RandomizedSearchCV(
    pipe_final, 
    param, 
    cv=4,
    scoring='f1', 
    n_jobs=-1,
    random_state=RANDOM_STATE
)
rand.fit(X_train, y_train)
rand.cv_results_

print('Лучшая модель и её параметры:\n\n', rand.best_estimator_)
print('Параметры лучшей модели:', rand.best_params_)
print('Метрика лучшей модели на тренировочной выборке:', rand.best_score_)
```

```python
f1_score(y_valid, rand.predict(X_valid))
```
